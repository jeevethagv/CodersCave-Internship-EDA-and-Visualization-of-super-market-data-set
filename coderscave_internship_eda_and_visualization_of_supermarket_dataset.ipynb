This project is an EDA of Super Market dataset. The dataset contains information regarding the sales of ten different stores during a period of three months. I have used libraries like Pandas, Plotly,  and some built-in EDA functions.

The line import pandas as pd is used to load the dataset.

```python
import plotly.express as px
import pandas as pd
data=pd.read_csv("/content/drive/MyDrive/Hackathon_Ideal_Data.csv")
```

The groupby() function is used to group the month and sum of quantity respective to month and its represented in bar chart. Month 2 has the highest sales in quantity of products compared to month 1&2.

```python
m=data.groupby('MONTH')['QTY'].sum().reset_index()
print(m)
fig = px.bar(m, x='MONTH', y='QTY', title='Total Quantity by Month')
fig.show()
```

The line import pandas as pd is used to load the dataset.

```python
import plotly.express as px
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Hackathon_Working_Data.csv")
```

The groupby() function is used to group the storecode and sum of price respective to month and its represented in histogram. The store N7 has the highest sum of price of all the stores.

```python
m=df.groupby('STORECODE')['PRICE'].sum().reset_index()
print(m)
fig = px.histogram(m, x=df['STORECODE'], y=df['PRICE'], title='Total Quantity by Store Code',color=df['STORECODE'])
fig.show()
```

The groupby() function is used to group the brand  and sum of price, quantity and  value respective to brand and its represented in horizontal bar chart. Amul holds the top position among all brands.

```python
top_brands_sales = df.groupby('BRD').sum().sort_values("VALUE", ascending=False)[['QTY', 'PRICE', 'VALUE']][0:25]

fig = px.bar(top_brands_sales, x='VALUE', y=top_brands_sales.index, orientation='h',
             labels={'VALUE': 'Sales', 'BRD': 'BRD'},
             title="Sales from Top 25 Brands",
             width=800, height=600,color = 'VALUE')
fig.show()
```


This code analyzes and visualizes total sales for different brands in a DataFrame. It filters brands with total sales over $100,000 and creates a horizontal bar chart using Plotly Express.

```python
unique_brands = df['BRD'].unique()
brand_sums = {}
for brand in unique_brands:
    brand_sums[brand] = df[df['BRD'] == brand]['BILL_AMT'].sum()
brand_sums_df = pd.DataFrame(list(brand_sums.items()), columns=['BRD', 'Total_Sales'])
brand_sums_df = brand_sums_df.sort_values('Total_Sales', ascending=True)
print(brand_sums_df)
filtered_brands = brand_sums_df[brand_sums_df['Total_Sales'] > 100000]
print(filtered_brands)
fig = px.bar(filtered_brands, x='Total_Sales', y='BRD', orientation='h', title='Total Sales by Brand (Filtered)',color='BRD')
fig.show()
```
